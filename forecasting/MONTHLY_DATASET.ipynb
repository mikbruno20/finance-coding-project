```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import datetime as dt
from datetime import timedelta
```

```python
demand_factors_monthly = pd.read_csv("demand factors/demand_factors_monthly.csv")
financial_factors_monthly = pd.read_csv("financial factors/monthly_financial_factors.csv")
implied_vol_indices_monthly = pd.read_csv("implied vol indices/monthly_implied_volatility_indices.csv")
brent_monthly = pd.read_csv("prices/monthly_realized_volatility_Brent.csv")
supply_factors_monthly = pd.read_csv("supply factors/monthly_supply_factors.csv")
uncertainty_factors_monthly = pd.read_csv("uncertainty factors/uncertainty factors monthly.csv")
```

```python
demand_factors_monthly = demand_factors_monthly.set_index("Date")
financial_factors_monthly = financial_factors_monthly.set_index("Date")
implied_vol_indices_monthly = implied_vol_indices_monthly.set_index("Date")
brent_monthly = brent_monthly.set_index("Date")
supply_factors_monthly = supply_factors_monthly.set_index("Date")
uncertainty_factors_monthly = uncertainty_factors_monthly.set_index("Date")
```

```python
monthly_dataset = pd.merge(demand_factors_monthly, financial_factors_monthly, on = "Date")
monthly_dataset = pd.merge(monthly_dataset,implied_vol_indices_monthly , on = "Date")
monthly_dataset = pd.merge(monthly_dataset, brent_monthly, on = "Date")
monthly_dataset = pd.merge(monthly_dataset, supply_factors_monthly, on = "Date")
monthly_dataset = pd.merge(monthly_dataset, uncertainty_factors_monthly, on = "Date")
```

```python
monthly_dataset
```

```python
y = pd.read_csv("prices/monthly_realized_volatility_WTI.csv")
y.set_index("Date", inplace=True)
```

```python
y = y.tail(206)
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    return {"Mean Squared Error": mse, "Mean Absolute Error": mae, "R^2 Score": r2}
```

```python
lin_model_ridge = Ridge(alpha=1.0)
lin_model_ridge.fit(monthly_dataset, y)
metrics_ridge = evaluate_model(lin_model_ridge, monthly_dataset, y)
metrics_ridge
```

```python
lin_model_lasso = Lasso(alpha=0.1)
lin_model_lasso.fit(monthly_dataset, y)
metrics_lasso = evaluate_model(lin_model_lasso, monthly_dataset, y)
metrics_lasso
```

```python
lin_model_elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
lin_model_elastic.fit(monthly_dataset, y)
metrics_elastic = evaluate_model(lin_model_elastic, monthly_dataset, y)
metrics_elastic
```

```python
X_train, X_test, y_train, y_test = train_test_split(monthly_dataset, y, test_size=0.2, random_state=42)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train.values.ravel())
metrics_rf = evaluate_model(rf_model, X_test, y_test)
metrics_rf
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
metrics_dt = evaluate_model(dt_model, X_test, y_test)
metrics_dt
```

```python
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
```

```python
svm_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svm_model.fit(X_train, y_train.values.ravel())
svm_metrics = evaluate_model(svm_model, X_test, y_test)
svm_metrics
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPRegressor
```

```python
model_ann = Pipeline([
    ('scaler', StandardScaler()),       
    ('ann', MLPRegressor(hidden_layer_sizes=(50, 25), activation='relu', solver='adam', max_iter=1000, random_state=42))
    ])
model_ann.fit(X_train, y_train.values.ravel())
ann_metrics = evaluate_model(model_ann, X_test, y_test)
ann_metrics
```
