```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from datetime import timedelta
```

```python
demand_factors_daily = pd.read_csv("demand factors/demand_factors_daily.csv")
financial_factors_daily = pd.read_csv("financial factors/daily_financial_factors.csv")
implied_vol_indices_daily = pd.read_csv("implied vol indices/daily_implied_volatility_indices.csv")
brent_daily = pd.read_csv("prices/daily_realized_volatility_Brent.csv")
supply_factors_daily = pd.read_csv("supply factors/daily_oil_data.csv")
uncertainty_factors_daily = pd.read_csv("uncertainty factors/uncertainty factors daily.csv")
```

```python
demand_factors_daily = demand_factors_daily.set_index("Date")
financial_factors_daily = financial_factors_daily.set_index("Date")
implied_vol_indices_daily = implied_vol_indices_daily.set_index("Date")
brent_daily = brent_daily.set_index("Date")
supply_factors_daily = supply_factors_daily.set_index("Date")
uncertainty_factors_daily = uncertainty_factors_daily.set_index("Date")
```

```python
daily_dataset = pd.merge(demand_factors_daily, financial_factors_daily, on = "Date")
daily_dataset = pd.merge(daily_dataset, implied_vol_indices_daily, on = "Date")
daily_dataset = pd.merge(daily_dataset, brent_daily, on = "Date")
daily_dataset = pd.merge(daily_dataset, supply_factors_daily, on = "Date")
daily_dataset = pd.merge(daily_dataset, uncertainty_factors_daily, on = "Date")
```

```python
daily_dataset = pd.concat([demand_factors_daily, financial_factors_daily])
daily_dataset = pd.concat([daily_dataset, implied_vol_indices_daily])
daily_dataset = pd.concat([daily_dataset, brent_daily])
daily_dataset = pd.concat([daily_dataset, supply_factors_daily])
daily_dataset = pd.concat([daily_dataset, uncertainty_factors_daily])
```

```python
daily_dataset = daily_dataset.fillna(0)
```

```python
daily_dataset = daily_dataset.iloc[64396:,:]
daily_dataset
```

```python
rev_daily = pd.read_csv("prices/daily_realized_volatility_WTI.csv")
rev_daily.set_index("Date", inplace = True)
```

```python
rev_daily
```

RIDGE

```python
from sklearn.linear_model import Ridge
```

```python
model_ridge = Ridge(max_iter=10000, alpha=1.0, solver="auto", random_state=42)
model_ridge.fit(daily_dataset, rev_daily)
model_ridge.predict(daily_dataset)
```

```python
model_ridge
```

LASSO

```python
from sklearn.linear_model import Lasso
```

```python
model_lasso = Lasso(max_iter=10000, alpha=0.1, random_state=42)
model_lasso.fit(daily_dataset, rev_daily)
model_lasso.predict(daily_dataset)
model_lasso
```

ELASTIC NET

```python
from sklearn.linear_model import ElasticNet
```

```python
model_elastic = ElasticNet(max_iter=10000, alpha=0.1, l1_ratio=0.5, random_state=42)
model_elastic.fit(daily_dataset, rev_daily)
model_elastic.predict(daily_dataset)
model_elastic
```
