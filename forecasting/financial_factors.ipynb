```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
from datetime import timedelta
import yfinance as yf
import seaborn as sns
import openpyxl
```

- oro  GC = F
- rame  HG = F
- grano  ZW = F

```python
ticker = ["^GSPC", "USDEUR=X", "JPY=X", "GBPUSD=X", "HG=F", "ZW=F", "GC=F"]

end = dt.datetime.now()
start = end - dt.timedelta(days = 365 * 45)


data = yf.download(ticker, start , end)["Close"].pct_change().dropna()
data
```

```python
daily_data = data
daily_data
```

```python
weekly_data = data.resample("W").mean()
weekly_data
```

```python
monthly_data = data.resample("M").mean()
monthly_data
```

```python
fed_funds = pd.read_excel("C:\\Users\\miche\\OneDrive\\Documenti\\PROJECTS\\crude oil forecasting\\financial_factors\\fed_funds_rate.xlsx", engine = "openpyxl")
```

```python
fed_funds.columns = ["Date", "Fed_Funds_Rate"]
fed_funds["Date"] = pd.to_datetime(fed_funds["Date"])
fed_funds.set_index("Date", inplace=True)
fed_funds
```

DAILY

```python
fed_funds_daily = fed_funds.resample("D").ffill().reset_index()
fed_funds_daily.set_index("Date", inplace = True)
fed_funds_daily = fed_funds_daily / 22 
fed_funds_daily
```

```python
data_daily = pd.merge(daily_data, fed_funds_daily, on = "Date")
```

```python
data_daily
```

```python
data_daily.to_csv('daily_financial_factors.csv')
```

WEEKLY

```python
fed_funds_week = fed_funds.resample("W").ffill().reset_index()
fed_funds_week.set_index("Date", inplace = True)
fed_funds_week = fed_funds_week / 4
fed_funds_week
```

```python
weekly_vol = pd.merge(weekly_data, fed_funds_week, on = "Date")
```

```python
weekly_vol
```

```python
weekly_vol.to_csv('weekly_financial_factors.csv')
```

MONTHLY

```python
fed_funds_month = fed_funds.resample("M").ffill().reset_index()
fed_funds_month.set_index("Date", inplace = True)
fed_funds_month
```

```python
month_vol = pd.merge(monthly_data, fed_funds_month, on = "Date")
```

```python
month_vol
```

```python
month_vol.to_csv('monthly_financial_factors.csv')
```
