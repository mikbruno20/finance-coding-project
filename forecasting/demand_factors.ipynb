```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import datetime as dt
from datetime import timedelta
```

```python
imports = pd.read_excel("C:\\Users\\miche\\OneDrive\\Documenti\\PROJECTS\\crude oil forecasting\\demand factors\\import_export.xls", sheet_name = "Data 1")
```

```python
ppi = pd.read_csv("C:\\Users\\miche\\OneDrive\\Documenti\\PROJECTS\\crude oil forecasting\\demand factors\\PPI_commodity.csv")
```

```python
consumption = pd.read_excel("C:\\Users\\miche\\OneDrive\\Documenti\\PROJECTS\\crude oil forecasting\\demand factors\\prod_supplied.xls", sheet_name = "Data 1")
```

IMPORTS

```python
imports = imports.iloc[:,:5]
imports.drop(imports.columns[1], axis = 1, inplace = True)
imports.drop(imports.columns[2], axis = 1, inplace = True)
imports
```

```python
imports.columns = ['Date', 'Imports_US', 'Imports_SPR']
imports = imports[~imports["Date"].isin(["Sourcekey", "Date"])].copy()
imports["Date"] = pd.to_datetime(imports["Date"])
imports.set_index("Date", inplace = True)
imports
```

```python
imports.fillna(0, inplace= True)
imports
```

```python
imports_weekly = imports.resample("W").mean()
imports_weekly.fillna(0, inplace = True)
```

```python
imports_daily = imports.resample("D").mean()
imports_daily.fillna(0, inplace = True)
```

PRODUCER PURCHASE INDEX

```python
ppi.columns = ["Date", "PPI_commodity"]
ppi["Date"] = pd.to_datetime(ppi["Date"])
ppi.set_index("Date", inplace = True)
ppi
```

```python
ppi_weekly = ppi.resample("W").mean()
ppi_weekly.fillna(0, inplace = True)
```

```python
ppi_daily = ppi.resample("D").mean()
ppi_daily.fillna(0, inplace = True)
```

CONSUMPTION

```python
consumption = consumption.iloc[:,:2]
consumption.columns = ["Date", "Consumption_US"]
```

```python
consumption = consumption.iloc[2: , : ]
consumption["Date"] = pd.to_datetime(consumption["Date"])
consumption.set_index("Date", inplace = True)
consumption
```

```python
consumption_monthly = consumption.resample("M").mean()
consumption_monthly.fillna(0, inplace = True)
```

```python
consumption_daily = consumption.resample("D").mean()
consumption_daily.fillna(0, inplace = True)
```

DAILY DATASET

```python
demand_factors_daily = pd.concat([imports_daily, ppi_daily, consumption_daily], axis = 1)
demand_factors_daily.fillna(0, inplace = True)
demand_factors_daily
```

```python
demand_factors_daily.to_csv("demand_factors_daily.csv")
```

WEEKLY DATASET

```python
demand_factors_weekly = pd.concat([imports_weekly, ppi_weekly, consumption], axis = 1)
demand_factors_weekly.fillna(0, inplace = True)
demand_factors_weekly
```

```python
demand_factors_weekly.to_csv("demand_factors_weekly.csv")
```

MONTHLY DATASET

```python
demand_factors_monthly = pd.concat([imports, ppi, consumption_monthly], axis = 1)
demand_factors_monthly.fillna(0, inplace = True)
demand_factors_monthly
```

```python
demand_factors_monthly.to_csv("demand_factors_monthly.csv")
```
