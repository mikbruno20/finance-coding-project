```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
from datetime import timedelta
```

```python
data = pd.read_excel("C:\\Users\\miche\\OneDrive\\Documenti\\PROJECTS\\crude oil forecasting\\uncertainty factors\\US_Policy_Uncertainty_Data.xlsx" )
```

```python
data.dropna(inplace = True)
data["Month"] = data["Month"].apply(lambda x : int(x))
data["Date"] = data["Year"].astype(str) + "-" + data["Month"].astype(str) + "-01"
data["Date"] = pd.to_datetime(data["Date"])
```

```python
data.drop(columns = ["Year", "Month"], inplace = True)
data.set_index("Date", inplace = True)
data.sort_index(ascending = False, inplace = True)
```

```python
data.columns = ["US Policy Uncertainty Index"]
```

```python
data
```

```python
import openpyxl

categorical = pd.read_excel("C:\\Users\\miche\\OneDrive\\Documenti\\PROJECTS\\crude oil forecasting\\uncertainty factors\\Categorical_EPU_Data.xlsx", engine = 'openpyxl') 
```

```python
categorical.dropna(inplace = True)
categorical["Month"] = categorical["Month"].apply(lambda x : int(x))
categorical["Date"] = categorical["Year"].astype(str) + "-" + categorical["Month"].astype(str) + "-01"
categorical["Date"] = pd.to_datetime(categorical["Date"])
categorical.drop(columns = ["Year", "Month"], inplace = True)
categorical.set_index("Date", inplace = True)
```

```python
categorical.drop(columns = ["1. Economic Policy Uncertainty"], inplace = True)
```

```python
categorical.columns = ["Monetary Policy", "Fiscal Policy", "Taxes", "Government Spending", "Health Care", "National Security", "Entitlement programs", "Regulation", "Financial Regulation", "Trade Policy", "Sovereign debt, currency crises"]
```

```python
categorical
```

MONTHLY

```python
uncert_index_month = data.resample("M").mean()
```

```python
categorical_month = categorical.resample("M").mean()
```

```python
uncertainty_factors_monthly = pd.merge(categorical_month, uncert_index_month, on = "Date")
```

```python
uncertainty_factors_monthly.to_csv("uncertainty factors monthly.csv")
```

WEEKLY

```python
uncert_index_week = data.resample("W").mean()
#uncert_index_week.dropna(inplace = True)
```

```python
uncert_index_week = uncert_index_week.apply(lambda x: x / 5).fillna(method='ffill')
```

```python
categorical_week = categorical.resample("W").mean()
categorical_week = categorical_week.apply(lambda x: x / 5).ffill()
categorical_week
```

```python
uncertainty_factors_weekly = pd.merge(categorical_week, uncert_index_week, on = "Date")
uncertainty_factors_weekly.to_csv("uncertainty factors weekly.csv")
```

DAILY

```python
uncertainy_index_daily = data.resample("D").mean()
```

```python
uncertainy_index_daily = uncertainy_index_daily.apply(lambda x: x / 22).fillna(method='ffill')
```

```python
uncertainy_index_daily
```
